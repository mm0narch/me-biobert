```python
from scipy import sparse
import numpy as np
import os
import joblib
from src.evaluation import evaluate_model
```

```python
tfidf_X_test=sparse.load_npz('../data/clean/tfidf_test.npz')
tfidf_y_test=np.load('../data/clean/tfidf_ytest.npz.npy')

biobert_X_test=np.load('../data/clean/biobert_test.npy')
biobert_y_test=np.load('../data/clean/biobert_ytest.npy')
```

```python
tfidf_models_dir='../models/saved/tfidf'
biobert_models_dir='../models/saved/biobert'
tfidf_model_files=os.listdir(tfidf_models_dir)
biobert_model_files=os.listdir(biobert_models_dir)

tfidf_models=[]
biobert_models=[]

for file in tfidf_model_files:
    tfidf_name=file.replace('.joblib', '')
    tfidf_model=joblib.load(os.path.join(tfidf_models_dir, file))
    tfidf_models.append({'name': tfidf_name, 'model': tfidf_model})

for file in biobert_model_files:
    biobert_name=file.replace('.joblib', '')
    biobert_model=joblib.load(os.path.join(biobert_models_dir, file))
    biobert_models.append({'name': biobert_name, 'model': biobert_model})

print(biobert_models)
```

```python
tfidf_model_evaluation=evaluate_model(tfidf_models, tfidf_X_test, tfidf_y_test)
```

```python
biobert_model_evaluation=evaluate_model(biobert_models, biobert_X_test, biobert_y_test)
```
