```python
from src.data_loader import load_data, save_file
from src.preprocessing import clean_chapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
```

```python
df_raw=load_data('../data/raw/icd10.csv')
df_clean, le=clean_chapter(df_raw)

df_clean.head()
```

```python
train_df, eval_df=train_test_split(
    df_clean,
    test_size=0.2,
    stratify=df_clean['Label'],
    random_state=42
)

val_df, test_df=train_test_split(
    eval_df,
    test_size=0.5,
    stratify=eval_df['Label'],
    random_state=42
)
```

```python
def plot_class_df_clean(df, label_col):
    plt.figure(figsize=(10,6))
    order=df[label_col].value_counts().index
    sns.countplot(data=df, x=label_col, order=order, palette='viridis')

    plt.xticks(rotation=90)
    plt.title('Class Distribution', fontsize=16)
    plt.xlabel('ICD10 Chapter')
    plt.ylabel('Sample Count')
    plt.tight_layout()
    plt.show()

plot_class_df_clean(df_clean, label_col='Kategori')
```

```python
def plot_class_train_df(df, label_col):
    plt.figure(figsize=(10,6))
    order=df[label_col].value_counts().index
    sns.countplot(data=df, x=label_col, order=order, palette='viridis')

    plt.xticks(rotation=90)
    plt.title('Class Distribution for Train Split', fontsize=16)
    plt.xlabel('ICD10 Chapter')
    plt.ylabel('Sample Count')
    plt.tight_layout()
    plt.show

plot_class_train_df(train_df, label_col='Kategori')
```

```python
def plot_class_val_df(df, label_col):
    plt.figure(figsize=(10,6))
    order=df[label_col].value_counts().index
    sns.countplot(data=df, x=label_col, order=order, palette='viridis')

    plt.xticks(rotation=90)
    plt.title('Class Distribution of Evaluation Split', fontsize=16)
    plt.xlabel('ICD10 Chapter')
    plt.ylabel('Sample Count')
    plt.tight_layout()
    plt.show

plot_class_val_df(val_df, label_col='Kategori')
```

```python
def plot_class_test_df(df, label_col):
    plt.figure(figsize=(10,6))
    order=df[label_col].value_counts().index
    sns.countplot(data=df, x=label_col, order=order, palette='viridis')

    plt.xticks(rotation=90)
    plt.title('Class Distribution of Test Split', fontsize=16)
    plt.xlabel('ICD10 Chapter')
    plt.ylabel('Sample Count')
    plt.tight_layout()
    plt.show()

plot_class_test_df(test_df, label_col='Kategori')
```

```python
save_file(train_df, '../data/clean/chaptertrain.csv')
save_file(val_df, '../data/clean/chaptervalidation.csv')
save_file(test_df, '../data//clean/chaptertest.csv')
```
